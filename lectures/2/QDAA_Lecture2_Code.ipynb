### **Handling Outliers**
One of the issues that can arise when we've gotten rid of all of the invalid data is some _really really really_ small or _really really really_ large numbers skewing our data. Sometimes they happen randomly. Othertimes don't belong in our dataset and we need to remove them.

There are a few different methods, but today we're going to consult statistics and use a formal method to determine what is an outlier and what isn't.

A given number n in a dataset is an outlier if

$ n < (q_1 - 1.5 \cdot IQR)$ \\
or
$ n > (q_3 + 1.5 \cdot IQR)$


*_Where_*

$ n = \text{ The number in our dataset }$ \\
$ IQR =  q_3 - q_1 \text{ Interquartile range, the difference between } q_3 \text{and }q_1$





#### Get our quartiles
Our quartiles are defined by the percentage of the way they are through the list if it were sorted. So quartile 0 (denoted as $q_0$) would be the minimum, or fist item in the list. $q_1$ is the 25th percentile, or item a quarter way down. $q_2$ is the 50th percentile, or median, and $q_3$ is the 75th percentile. $q_4$ is the maximum. Typically, $q_0$ and $q_4$ are referred to as min and max, and $q_2$ as the median. In practice, you will usually only see $q_1$ and $q_3$, but its handy to know what they really mean.

```python
q1 = df['Salary'].quantile(0.25)
q3 = df['Salary'].quantile(0.75)

q1, q3
```

```python
iqr = q3 - q1
upper_bound = q3 + 1.5*iqr
lower_bound = q1 - 1.5*iqr
print(lower_bound, upper_bound)
```

### **Now we can remove the outliers using.... logical arrays!**

```python
df = df[(df['Salary'] < upper_bound) & (df['Salary'] > lower_bound)]
```

```python
df.hist('Salary', bins=25)
```
